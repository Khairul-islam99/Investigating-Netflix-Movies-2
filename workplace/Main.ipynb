```python
import pandas as pd
import seaborn as sns
```

```python
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
#check Dataframe missing values.
netflix_df.isnull()
```

```python
#Find the top contry based on Movie and Tv show
top_country = netflix_df["country"].value_counts().idxmax()
top_country
```

```python
# Top 10 countrys based on TV shows
netflix_df["tv_show"] = netflix_df["type"]=="TV Show"
Tv_Shows_country= netflix_df[netflix_df["tv_show"]==True].value_counts([netflix_df["country"]]).head(10)
print(Tv_Shows_country)
```

```python
# Top 10 countrys  based on Movies
netflix_df["Movies"] = netflix_df["type"]=="Movie"
Movie_country= netflix_df[netflix_df["Movies"]==True].value_counts([netflix_df["country"]]).head(10)
print(Movie_country)
```

```python
# Top 10 genre 
top_genres = netflix_df["genre"].value_counts().head(10)
print(top_genres)
```

```python
#first relase date
first_relase_date =netflix_df["release_year"].sort_values(ascending= True).min()
print(first_relase_date)
```

```python
#Details of the first Movie relase on netflix
netflix_subset_movie = netflix_df[netflix_df["type"] == "Movie"]
first_relase_movie =netflix_subset_movie[(netflix_subset_movie["release_year"] == 1942)].max()
print(first_relase_movie)
```

```python
#top cast name 
top_cast = netflix_df["cast"].value_counts().head(1)
print(top_cast)
```
